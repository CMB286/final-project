# The Data:

The first step was to import the data by reading from the csv. We then have to group it by nieghborhood, sum the amounts together and sort the values.

```python
smart_trash = pd.read_csv("https://data.wprdc.org/datastore/dump/75b83ac9-8069-4cf1-bcc3-b9e6b04487d9")
smart_trash_grouped = smart_trash.groupby('neighborhood').size()
smart_trash_sorted = smart_trash_grouped.sort_values()
smart_trash_sorted.head(20)
```

```python
firearm_seizures = pd.read_csv("https://data.wprdc.org/datastore/dump/e967381d-d7e9-48e3-a2a2-39262f7fa5c4")
firearm_grouped = firearm_seizures.groupby('neighborhood').sum()
firearm_sorted = firearm_grouped.sort_values(by='total_count')
firearm_sorted['total_count'].head(20)
```

We then put each dataset into a dictionary to do some further analysis.

```python
trash_dict = smart_trash_sorted.to_dict()
trash_dict
```

```python
firearms_dict = firearm_sorted['total_count'].to_dict()
firearms_dict
```

# Analysis

In order to find the best neighborhood we must combine the firearm data and the trash data. This is done by subtracting the amount of trash cans from the amount of firearm siezures. This is because the large amount of trash cans is a positve influence while a large amount of firearm siezures is a negative influence.

```python
dict2 = {}

for key in firearms_dict.keys():
    if key in trash_dict:
        dict2[key] = (firearms_dict[key] - trash_dict[key]) 

sorted_values = sorted(dict2.items(), key = lambda kv:(kv[1], kv[0])) 

sorted_values
```
